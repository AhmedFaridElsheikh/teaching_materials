### 📋 CHEATSHEET — Syntax Reminders

| What you want | Syntax |
|---|---|
| First 5 rows | `df.head()` |
| Shape (rows, cols) | `df.shape` |
| Column names | `df.columns` |
| Data types + nulls | `df.info()` |
| Summary statistics | `df.describe()` |
| Count of each category | `df['col'].value_counts()` |
| Unique values | `df['col'].unique()` |
| Number of unique values | `df['col'].nunique()` |
| Mean of a column | `df['col'].mean()` |
| Filter rows | `df[df['col'] > value]` |
| Filter by text | `df[df['col'] == 'text']` |
| Group and aggregate | `df.groupby('col')['col2'].mean()` |
| Sort values | `df.sort_values('col', ascending=False)` |
| Check missing values | `df.isnull().sum()` |
| Select one column | `df['col']` |
| Select multiple columns | `df[['col1', 'col2']]` |
